## Calculando o comprimento da pista usando os tempos do arquivo "laps(arquivo3).xlsx", em metros(m):
Observa-se que o resultado é aproximadamente igual aos valores obtidos nos outros arquivos.

```python
distance1 = df.loc[df.time==163.02,'distance'].item()-df.loc[df.time==53.60,'distance'].item()
distance2 = df.loc[df.time==264.64,'distance'].item()-df.loc[df.time==163.02,'distance'].item()
distance3 = df.loc[df.time==366.93,'distance'].item()-df.loc[df.time==264.64,'distance'].item()
distance4 = df.loc[df.time==465.31,'distance'].item()-df.loc[df.time==366.93,'distance'].item()
distance5 = df.loc[df.time==561.67,'distance'].item()-df.loc[df.time==465.31,'distance'].item()
display(distance1)
display(distance2)
display(distance3)
display(distance4)
display(distance5)
distance = (distance1+distance2+distance3+distance4+distance5)/5
print('O comprimento da pista é:','%.2f' % distance,'metros')
```

## Tempo aproximado para cada volta, baseado na tabela "laps(arquivo3).xlsx", em segundos(s):

```python
first_lap = 163.02 - 53.60
second_lap = 264.64 - 163.02
third_lap = 366.93 - 264.64
fourth_lap = 465.31 - 366.93
fifth_lap = 561.67 - 465.31

print('Primeira volta:','%.2f' % first_lap, 's')
print('Segunda volta:','%.2f' % second_lap, 's')
print('Terceira volta:','%.2f' % third_lap, 's')
print('Quarta volta:','%.2f' % fourth_lap, 's')
print('Quinta volta:','%.2f' % fifth_lap, 's')
```

## Calculando a velocidade média usando a coluna horizontalSpeed, em metros por segundo(m/s):
Para isso, iremos tratar a tabela primeiro para remover tanto os intervalos nos quais o carro estava parado, quanto a volta que o piloto deu para retornar a posição inicial. Para o cálculo da velocidade média, iremos usar as 5 voltas dadas pelo piloto.

```python
#df = df.dropna()
df = df.drop_duplicates(['latitude','longitude'])
df = df.drop(df.loc[:, 'numSVsTracked':'wheelAccelRR'],axis=1)
df = df.drop(df[df.time<53.60].index,axis=0)
df = df.drop(df[df.time>561.67].index,axis=0)
df = df.reset_index(drop=True)
df
```

#### Verificando se a velocidade está com o tipo correto:

```python
df.horizontalSpeed.info()
```

```python
first_lap_AverageV = distance1/first_lap
first_lap_maxV = df.loc[(df.time>53.60) & (df.time<163.02),'horizontalSpeed'].max()

second_lap_AverageV = distance2/second_lap
second_lap_maxV = df.loc[(df.time>163.02) & (df.time<264.64),'horizontalSpeed'].max()

third_lap_AverageV = distance3/third_lap
third_lap_maxV = df.loc[(df.time>264.64) & (df.time<366.93),'horizontalSpeed'].max()

fourth_lap_AverageV = distance4/fourth_lap
fourth_lap_maxV = df.loc[(df.time>366.93) & (df.time<465.31),'horizontalSpeed'].max()

fifth_lap_AverageV = distance5/fifth_lap
fifth_lap_maxV = df.loc[(df.time>465.31) & (df.time<561.67),'horizontalSpeed'].max()



print('A velocidade média da primeira volta foi de:','%.2f' % first_lap_AverageV,'m/s')
print('A velocidade máxima da primeira volta foi de:','%.2f' % first_lap_maxV,'m/s')
print('\n')
print('A velocidade média da segunda volta foi de:','%.2f' % second_lap_AverageV,'m/s')
print('A velocidade máxima da segunda volta foi de:','%.2f' % second_lap_maxV,'m/s')
print('\n')
print('A velocidade média da terceira volta foi de:','%.2f' % third_lap_AverageV,'m/s')
print('A velocidade máxima da terceira volta foi de:','%.2f' % third_lap_maxV,'m/s')
print('\n')
print('A velocidade média da quarta volta foi de:','%.2f' % fourth_lap_AverageV,'m/s')
print('A velocidade máxima da quarta volta foi de:','%.2f' % fourth_lap_maxV,'m/s')
print('\n')
print('A velocidade média da quinta volta foi de:','%.2f' % fifth_lap_AverageV,'m/s')
print('A velocidade máxima da quinta volta foi de:','%.2f' % fifth_lap_maxV,'m/s')
```
