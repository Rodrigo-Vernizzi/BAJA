```python
import matplotlib as mpl
from matplotlib.collections import LineCollection
colormap = mpl.cm.plasma
import numpy as np 
```

```python
fd = pd.read_csv('20140221_01_02_03_250lm.csv')

x = fd['latitude']
y = fd['longitude']
color = fd.handwheelAngle

points = np.array([x, y]).T.reshape(-1, 1, 2)
segments = np.concatenate([points[:-1], points[1:]], axis=1)

# We create a plot with title and adjust some setting to make it look good.
fig, ax = plt.subplots(sharex=True, sharey=True, figsize=(12, 6.75))

# Adjust margins and turn of axis
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.12)
ax.axis('off')


# After this, we plot the data itself.
# Create background track line
ax.plot(fd['longitude'], fd['latitude'], color='black', linestyle='-', linewidth=16, zorder=0)

# Create a continuous norm to map from data points to colors
norm = plt.Normalize(color.min(), color.max())
lc = LineCollection(segments, cmap=colormap, norm=norm, linestyle='-', linewidth=5)

# Set the values used for colormapping
lc.set_array(color)

# Merge all line segments together
line = ax.add_collection(lc)


# Finally, we create a color bar as a legend.
cbaxes = fig.add_axes([0.25, 0.05, 0.5, 0.05])
normlegend = mpl.colors.Normalize(vmin=color.min(), vmax=color.max())
legend = mpl.colorbar.ColorbarBase(cbaxes, norm=normlegend, cmap=colormap, orientation="horizontal")


# Show the plot
plt.show()
```

## Calculando o comprimento da pista usando os tempos do arquivo "laps(arquivo2).xlsx", em metros(m):
Observa-se que o resultado é aproximadamente igual aos valores obtidos nos outros arquivos.

```python
distance1 = df.loc[df.time==361.35,'distance'].item()-df.loc[df.time==250.40,'distance'].item()
distance2 = df.loc[df.time==469.65,'distance'].item()-df.loc[df.time==361.35,'distance'].item()
distance3 = df.loc[df.time==572.58,'distance'].item()-df.loc[df.time==469.65,'distance'].item()
display(distance1)
display(distance2)
display(distance3)
display(df.loc[df.time==572.58,'distance'].item())
distance = (distance1+distance2+distance3)/3
print('O comprimento da pista é:','%.2f' % distance,'metros')
```

## Tempo aproximado para cada volta, baseado na tabela"primeiravolta(arquivo2).xlsx", em segundos(s):

```python
first_lap = 361.35 - 250.40
second_lap = 469.65 - 361.35
third_lap = 572.58 - 469.65

print('Primeira volta:','%.2f' % first_lap, 's')
print('Segunda volta:','%.2f' % second_lap, 's')
print('Terceira volta:','%.2f' % third_lap, 's')
```

## Calculando a velocidade média usando a coluna horizontalSpeed, em metros por segundo(m/s):
Para isso, iremos tratar a tabela primeiro para remover tanto os intervalos nos quais o carro estava parado, quanto a volta que o piloto deu para retornar a posição inicial. Para o cálculo da velocidade média, iremos usar as 3 voltas dadas pelo piloto.

```python
df = df.dropna()
#tabel[['latitude','longitude']].drop_duplicates()
df = df.drop_duplicates(['latitude','longitude'])
df = df.drop(df.loc[:, 'numSVsTracked':'wheelAccelRR'],axis=1)
df = df.drop(df[df.time<250.40].index,axis=0)
df = df.drop(df[df.time>572.58].index,axis=0)
df = df.reset_index(drop=True)
df
```

#### Verificando se a velocidade está com o tipo correto:

```python
df.horizontalSpeed.info()
```

```python
first_lap_AverageV = distance1/first_lap
first_lap_maxV = df.loc[(df.time>250.40) & (df.time<361.35),'horizontalSpeed'].max()

second_lap_AverageV = distance2/second_lap
second_lap_maxV = df.loc[(df.time>361.35) & (df.time<469.65),'horizontalSpeed'].max()

third_lap_AverageV = distance3/third_lap
third_lap_maxV = df.loc[(df.time>469.65) & (df.time<572.58),'horizontalSpeed'].max()

print('A velocidade média da primeira volta foi de:',first_lap_AverageV,'m/s')
print('A velocidade máxima da primeira volta foi de:',first_lap_maxV,'m/s')
print('\n')
print('A velocidade média da segunda volta foi de:',second_lap_AverageV,'m/s')
print('A velocidade máxima da segunda volta foi de:',second_lap_maxV,'m/s')
print('\n')
print('A velocidade média da terceira volta foi de:',third_lap_AverageV,'m/s')
print('A velocidade máxima da terceira volta foi de:',third_lap_maxV,'m/s')
```
